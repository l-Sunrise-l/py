## Задача 2. Карма
Один буддист-программист решил создать свой симулятор жизни, в котором нужно набрать определённое 500 очков кармы (это константа), чтобы достигнуть просвещения. 

Каждый день вызывается специальная функция one_day() которая возвращает количество кармы от 1 до 7 и может с вероятностью 1 к 10 выкинуть одно из исключений:
- KillError
- DrunkError
- CarCrashError
- GluttonyError
- DepressionError

Напишите такую программу. Функцию оберните в бесконечный цикл, выход из которого возможен только при накоплении кармы до уровня константы. Исключения обработайте и запишите в отдельный лог karma.log




# Решение:

```python
from random import randint, choice


class KillError(Exception):
    def __str__(self):
        return '\nKillError'


class DrunkError(Exception):
    def __str__(self):
        return '\nDrunkError'


class CarCrashError(Exception):
    def __str__(self):
        return '\nCarCrashError'


class GluttonyError(Exception):
    def __str__(self):
        return '\nGluttonyError'


class DepressionError(Exception):
    def __str__(self):
        return '\nDepressionError'


listt = [DepressionError, CarCrashError, KillError, GluttonyError, DrunkError]
karma = 0
while karma < 500:
    try:
        if randint(1, 10) == 10:
            raise choice(listt)
        karma_1 = randint(1, 7)
        karma += karma_1
        print('Вы сегодня получили', karma_1, 'кармы', '\nСейчас у', karma, 'кармы')
    except (DepressionError, CarCrashError, KillError, GluttonyError, DrunkError) as err:
        data = open('karma.txt', 'a')
        data.write('\n' + str(err))
        data.close()
        print(err)
```

# Решение:

```python
class Person():

    def __init__(self, name, surname, age):
        self.name = name
        self.surname = surname
        self.age = age
        self.salary = 0

    def get_info(self):
        print(self.name)
        print(self.surname)
        print(self.age)


class Manager(Person):

    def __init__(self, name, surname, age):
        super().__init__(name, surname, age)
        self.salary = 13000

    def get_salary(self):
        return self.salary


class Agent(Person):

    def __init__(self, name, surname, age, volume_of_sales):
        super().__init__(name, surname, age)
        self.salary = 5000 + volume_of_sales * 0.05

    def get_salary(self):
        return self.salary


class Worker(Person):

    def __init__(self, name, surname, age, hours):
        super().__init__(name, surname, age)
        self.salary = 100 * hours

    def get_salary(self):
        return self.salary


manager_1 = Manager('Орлова', 'Елизавета', 30)
manager_2 = Manager('Соловьев', 'Иван', 30)
manager_3 = Manager('Клюева', 'Ольга', 31)
agent_1 = Agent('Иванова', 'Мадина', 29, 100000)
agent_2 = Agent('Николаев', 'Богдан', 29, 120000)
agent_3 = Agent('Зайцева', 'Алиса', 31, 140000)
worker_1 = Worker('Шубина', 'Дарья', 30, 100)
worker_2 = Worker('Николаева', 'Алина', 31, 120)
worker_3 = Worker('Суркова', 'Тамара', 29, 140)
listt = [manager_1, manager_2, manager_3, agent_1, agent_2, agent_3, worker_1, worker_2, worker_3]
summ = 0
for counter in listt:
    summ += counter.salary
print('заработная плата всех:', int(summ))
```

## Задача 5. А-а-автомобиль!
Автомобиль имеет координаты своего положения и угол, описывающий направление движения. Он может быть изначально 
поставлен в любую точку с любым направлением (конструктор), может проехать в выбранном направлении определенное 
расстояние и может повернуть, то есть изменить текущее направление на любое другое (передаём привет математике и формулам)

Реализуйте класс Автомобиля, а также класс, который будет описывать автобус. Кроме того, что имеется у автомобиля,
у автобуса должны быть поля, содержащие число пассажиров и количество полученных денег, изначально равные нулю. 
Также, должны быть методы «войти» и «выйти», изменяющие число пассажиров. Наконец, метод move должен быть переопределён,
чтобы увеличивать количество денег в соответствии с количеством пассажиров и проезжающим расстоянием.




# Решение:

```python
from math import cos, sin, radians


class Car():

    def __init__(self):
        self.cord_x = 0
        self.cord_y = 0

    def info(self):
        print('Координаты x:', self.cord_x)
        print('Координаты y:', self.cord_y)

    def move(self, degree, distance):
        self.cord_x += round(distance * cos(radians(degree)), 2)
        self.cord_y += round(distance * sin(radians(degree)), 2)
        self.info()


class Bus(Car):

    def __init__(self, ):
        super().__init__()
        self.money = 0
        self.passengers = 0

    def add_passenger(self, passengers):
        self.passengers += passengers
        print('Прибавилось', passengers, 'пассажиров')

    def remove_passenger(self, passengers):
        self.passengers -= passengers
        print('Убавилось', passengers, 'пассажиров')

    def info(self):
        print('Координаты x:', self.cord_x)
        print('Координаты y:', self.cord_y)
        print('Кол-во пассажир:', self.passengers)
        print('Кол-во денег:', self.money)

    def move(self, degree, distance):
        self.cord_x += round(distance * cos(radians(degree)), 2)
        self.cord_y += round(distance * sin(radians(degree)), 2)
        self.money += self.passengers * distance * 0.5
        self.info()


car = Car()
bus = Bus()
car.move(60, 100)
bus.add_passenger(12)
bus.remove_passenger(2)
bus.move(30, 100)
```

## Задача 7. Стек
Мы уже говорили, что в программировании нередко необходимо создавать свои собственные структуры данных на основе уже существующих. Одним из таких “базовых” структур является стек. 
Стек - это абстрактный тип данных, представляющий собой список элементов, организованных по принципу LIFO (англ. last in — first out, «последним пришёл — первым вышел»).

Простой пример: стек из книг на столе. Единственной книгой, чья обложка видна, является самая верхняя. Чтобы получить доступ к, например, третьей снизу книге, нам нужно убрать все книги, лежащие сверху, одну за другой.

Напишите класс, который реализует Стек и его возможности (достаточно будет добавления и удаления элемента). 

После этого напишите ещё один класс “Менеджер задач”. В менеджере задач можно выполнить команду “новая задача”, в которую передаётся сама задача (str) и её приоритет (int). Сам менеджер работает на основе Стэка (не наследование!).  При выводе менеджера в консоль все задачи должны быть отсортированы по приоритету: чем меньше число, тем выше задача.

Вот пример основной программы:

# Решение:

```python
class Stack():
    def __init__(self):
        self.listt = []

    def __str__(self):
        return self.listt

    def check(self):
        if len(self.listt) == 0:
            return True
        else:
            return False

    def new_task(self, word):
        self.listt.append(word)

    def remove_task(self, word):
        if word in self.listt:
            self.listt.remove(word)


class TaskManager():

    def __init__(self):
        self.listt = {}

    def new_task(self, word, number):
        if not number in self.listt:
            self.listt[number] = Stack()
        self.listt[number].new_task(word)

    def remove_task(self, word, number):
        self.listt[number].remove_task(word)
        if self.listt[number].check():
            self.listt.pop(number)

    def __str__(self):
        answer_1 = sorted(self.listt.items(), key=lambda x: x)
        answer = ''
        for counter in answer_1:
            answer += '\n' + str(counter[0])
            for counter_1 in counter[1].listt:
                answer += ' ' + counter_1 + ';'

        return answer


manager = TaskManager()

manager.new_task("сделать уборку", 4)
manager.new_task("помыть посуду", 4)
manager.new_task("отдохнуть", 1)
manager.new_task("поесть", 2)
manager.new_task("сдать дз", 2)
print(manager)

manager.remove_task('отдохнуть', 1)
manager.remove_task('поесть', 2)
manager.remove_task('спать', 4)
print(manager)
```
